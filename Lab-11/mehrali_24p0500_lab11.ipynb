## 1.1 What is Unsupervised Learning?
In previous labs, we worked on supervised learning methods like KNN and Naive Bayes, where correct labels were already available. In unsupervised learning, labels are not given. We only have raw data, and the model tries to find hidden structure by itself.

A simple example is organizing LEGO pieces without instructions. We can group them by color, size, or shape. In the same way, unsupervised learning groups data based on similarity.

## 1.2 What is Clustering?
Clustering is the most common unsupervised learning task. It divides data into groups (clusters) so that points in the same cluster are similar, and points in different clusters are different.

A real-life example is a supermarket arrangement: dairy in one area, produce in another, and cleaning items in a separate section.

## 1.3 The K-Means Algorithm
K-Means is a simple and widely used clustering algorithm. Here, **K** is the number of clusters, and each cluster is represented by a center point called a **centroid** (mean).

The goal is to place K centroids such that each point is as close as possible to its nearest centroid.

### 1.3.1 How It Works (Step by Step)
1. **Initialization:** Choose K and select K starting centroids.
2. **Assignment Step:** Assign every point to the nearest centroid.
3. **Update Step:** Recalculate each centroid as the mean of points in that cluster.
4. **Repeat:** Continue steps 2 and 3 until centroids stop moving.

## 1.4 Distance and Centroids
### 1.4.1 Euclidean Distance
K-Means uses Euclidean distance to find the nearest centroid:

\[
d = \sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2}
\]

### 1.4.2 Centroid (Mean) Calculation
For a cluster, centroid is found by averaging each feature value of all points in that cluster.

## 1.6 How to Choose K? (Elbow Method)
To select a good K, we use the Elbow Method. We run K-Means with different K values and calculate SSE (Sum of Squared Errors) each time.

\[
\mathrm{SSE} = \sum_{i=1}^{K} \sum_{x \in C_i} \mathrm{distance}(x, \mu_i)^2
\]

As K increases, SSE decreases. The best K is usually at the "elbow" point where the decrease becomes slower.

```python
import math
import random

# A simple 2D dataset.
# We can clearly see two groups visually:
# (1,1), (1,2) vs (10,10), (10,11)
dataset = [
    [1, 1], [1, 2], [2, 1],
    [10, 10], [10, 11], [11, 10]
]

def euclidean_distance(point1, point2):
    total = 0
    for i in range(len(point1)):
        diff = point1[i] - point2[i]
        total = total + (diff * diff)
    return math.sqrt(total)
```

```python
def get_closest_centroid(point, centroids):
    # Start with a very large distance (infinity) so the first distance we
    # check will always be smaller
    min_dist = float('inf')
    closest_index = -1

    for i in range(len(centroids)):
        dist = euclidean_distance(point, centroids[i])
        if dist < min_dist:
            min_dist = dist
            closest_index = i

    return closest_index
```

```python
def update_centroids(clusters):
    new_centroids = []

    for cluster in clusters:
        # If a cluster randomly ended up empty, keep it empty to avoid divide-by-zero
        if len(cluster) == 0:
            continue

        num_points = len(cluster)

        # Calculate the average across all dimensions (X, Y)
        new_centroid = []
        for dim in range(len(cluster[0])):
            dim_sum = 0
            for point in cluster:
                dim_sum += point[dim]
            new_centroid.append(dim_sum / num_points)

        new_centroids.append(new_centroid)

    return new_centroids
```

```python
def k_means(dataset, k, max_iterations=10):
    # Step 1: Initialize random centroids
    # Pick the first 'k' points as a starting point (simple approach)
    centroids = dataset[:k]

    for iteration in range(max_iterations):
        print(f"--- Iteration {iteration + 1} ---")

        # Create empty lists for our k clusters
        clusters = []
        for _ in range(k):
            clusters.append([])

        # Step 2: Assign points
        for point in dataset:
            centroid_index = get_closest_centroid(point, centroids)
            clusters[centroid_index].append(point)

        # Step 3: Update centroids
        new_centroids = update_centroids(clusters)

        # Check if they stopped moving
        if new_centroids == centroids:
            print("Centroids stopped moving! Convergence reached.")
            break

        centroids = new_centroids
        print("Centroids updated to:", centroids)

    return centroids, clusters

# Run the algorithm
k = 2
final_centroids, final_clusters = k_means(dataset, k)

for i in range(len(final_clusters)):
    print(f"Cluster {i + 1} points:", final_clusters[i])
```

## Task 1: Calculate the Error (SSE)

```python
def calculate_sse(clusters, centroids):
    total_error = 0
    # Loop through every cluster
    for i in range(len(clusters)):
        # Loop through every point in that cluster
        for point in clusters[i]:
            # Calculate the squared euclidean_distance between point and centroid
            error = euclidean_distance(point, centroids[i]) ** 2
            # Add it to total_error
            total_error += error
    return total_error

sse_value = calculate_sse(final_clusters, final_centroids)
print("SSE:", sse_value)
```

## Task 2: Predict New Data

```python
def predict(new_point, final_centroids):
    # Return the index of the closest centroid
    return get_closest_centroid(new_point, final_centroids)

print("Point [8, 8] belongs to cluster:", predict([8, 8], final_centroids))
```
