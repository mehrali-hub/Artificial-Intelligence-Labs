## Task 3
Plot two lines and set line markers.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 11)

y1 = 2*x + 1
y2 = np.cos(x)

plt.plot(x, y1, marker='o', label='y = 2x + 1')
plt.plot(x, y2, marker='x', label='cos')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Straight line and Cosine")

plt.legend()

plt.show()
```

## Task 4
Change the datatype of a given Series.

```python
import pandas as pd

data = pd.Series([100,200,300,400])

print("Original Data Series:")
print(data)

data = data.astype('float64')

print("After changing datatype:")
print(data)
```

## Task 5
Delete DataFrame rows based on a given column value.

```python
import pandas as pd

data = {
    'col1':[1,2,3,4,4],
    'col2':[4,5,6,7,8],
    'col3':[7,8,9,0,1]
}

df = pd.DataFrame(data)

print("Original DataFrame")
print(df)

new_df = df[df.col1 != 3]

print("New DataFrame")
print(new_df)
```
